```python
weather_mar2012 = pd.read_csv('weather_march_2012_hours.csv', skiprows=15, index_col='Date/Time', parse_dates=True, encoding='latin1')
```

```python
weather_mar2012.columns
```

```python
weather_mar2012["Wind Spd (km/h)"].plot()
```

```python
weather_mar2012.columns=['Year', 'Month', 'Day', 'Time', 'DataQuality', 'Temp',
       'TempFlag', 'DewPointTemp', 'DewPointTempFlag',
       'RelHum', 'RelHumFlag', 'WindDir', 'WindDirFlag',
       'WindSpd', 'WindSpdFlag', 'Visibility',
       'VisibilityFlag', 'StnPress', 'StnPressFlag', 'Hmdx',
       'HmdxFlag', 'WindChill', 'WindChillFlag', 'Weather']
```

```python
weather_mar2012.columns
```

```python
weather_mar2012.dropna(axis=1, how='all')
```

```python
weather_mar2012.drop(['Year', 'Month', 'Day', 'Time', 'DataQuality'],axis=1)
```

```python
WindSpeed = weather_mar2012[['WindSpd']].copy()
```

```python
WindSpeed.head()
```

```python
weather_mar2012.index.hour
```

```python
#temperatures.loc[:,'Hour']
WindSpeed.loc[:,'Hour'] = weather_mar2012.index.hour
```

```python
WindSpeed.head()
```

```python
mean_wind = WindSpeed.groupby('Hour').aggregate(np.mean)
```

```python
mean_wind
```

```python
mean_wind.plot()
```
