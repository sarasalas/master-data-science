```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15,5)

```

```python
complaints = pd.read_csv('311-service-requests.csv')
```

```python
complaints[:5]
```

```python
#seleccionamos las quejas por ruido
noise_complaints = complaints[complaints['Complaint Type'] == 'Noise - Street/Sidewalk']
```

```python
noise_complaints
```

```python
# vamos a combinar las condiciones: quejas por ruido en Brooklyn. Aqui definimos el array booleano que luego usamos
# en la condiciÃ³n
is_noise = complaints['Complaint Type'] == 'Noise - Street/Sidewalk'
in_Brooklyn = complaints['Borough'] == 'BROOKLYN'
complaints[is_noise & in_Brooklyn][:5]
```

```python
#seleccionamos algunas columnas con registros que cumplen la condiciÃ³n
complaints[is_noise & in_Brooklyn][['Complaint Type', 'Borough', 'Created Date']]
```

```python
#seleccionamos los registros que son por ruido
noise_complaints = complaints[is_noise]
```

```python
# del df generado seleccionamos la columna Borough y contamos los registros distintos
noise_complaints['Borough'].value_counts()
```

```python
# Calcular la proporciÃ³n de quejas por por ruido respecto del total de quejas (de todo tipo) en cada borough
```

```python
noise_complaint_count = noise_complaints['Borough'].value_counts()
noise_complaint_count
```

```python
complaints_count = complaints['Borough'].value_counts()
complaints_count
```

```python
# la proporciÃ³n es: quejas por ruido / total quejas
noise_complaint_prop = noise_complaint_count / complaints_count
noise_complaint_prop
```
